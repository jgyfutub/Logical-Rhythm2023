```python
import os
import tensorflow as tf
from tensorflow.keras.preprocessing.image import ImageDataGenerator
from tensorflow.keras.applications import MobileNetV2
from tensorflow.keras import models, layers
from tensorflow.keras.models import Sequential
```

```python
classes=os.listdir("/kaggle/input/anyas-gojo-revival-aerospace-enigma/train")
print(classes)
```

```python
train_dir="/kaggle/input/anyas-gojo-revival-aerospace-enigma/train"
IMG_SIZE=(256,256)
BATCH_SIZE=32
```

```python
train_augment=ImageDataGenerator(shear_range=0.2,zoom_range=0.2,validation_split=0.20,featurewise_center=True,rotation_range=0.05,width_shift_range=0.05,horizontal_flip=True)
train_data = train_augment.flow_from_directory(train_dir,target_size=IMG_SIZE,batch_size=BATCH_SIZE,seed=42,subset = 'training',shuffle=True,class_mode='categorical')
# validation_data = train_augment.flow_from_directory(train_dir,target_size=IMG_SIZE,batch_size=BATCH_SIZE,seed=42,subset = 'validation',shuffle=True,class_mode='categorical')
```

```python
train_data.class_indices
```

```python
base_model = MobileNetV2(weights='imagenet', include_top=False)
```

```python
model=Sequential([
    layers.Rescaling(1./255),
    base_model,
    layers.Flatten(),
    layers.Dense(5,activation='sigmoid')
])
```

```python
model.compile(
    optimizer='adam',
    loss='binary_crossentropy',
    metrics=['accuracy'],
)

model.fit(
    train_data,
#     validation_data=validation_data,
    epochs=10,
)
```
